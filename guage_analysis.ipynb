```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("Data/Daily_Max_Vals.csv")
```

```python
dates = df['date'].to_numpy()
year = []

for date in dates:
    year.append(date[0:4])

df["year"] = year
```

```python
years = df["year"].unique()
```

```python
yearly_guage = []

for year in years:
    temp_df = df[df['year'] == year]
    yearly_guage.append(temp_df["gauge"].mean())

yearly_guage
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
ax = sns.lineplot(x=years, y=yearly_guage)
ax.set_xticklabels(years, rotation=30)
```
